```python
import pandas as pd
import streamlit as st

DATA_PATH = r"C:\Users\prana\Downloads\netflix engagement intelligence\data"

df = pd.read_csv(DATA_PATH + r"\final_user_intelligence_table.csv")

st.set_page_config(page_title="Netflix Engagement Intelligence", layout="wide")

st.title("🎬 Netflix Engagement Intelligence Platform")

user_id = st.selectbox("Select User ID", df["user_id"].unique())

user = df[df["user_id"] == user_id].iloc[0]

st.subheader("User Risk Summary")

col1, col2, col3 = st.columns(3)

col1.metric("Risk Probability", f"{user['risk_probability']:.2%}")
col2.metric("Risk Level", user["risk_level"])
col3.metric("Subscription Plan", user["subscription_plan"])

st.subheader("Explanation")
st.write(user["risk_explanation"])

st.subheader("Recommended Business Action")
st.success(user["recommended_action"])

st.subheader("User Profile")

profile_cols = [
    "age", "gender", "country", "state_province",
    "monthly_spend", "primary_device", "household_size"
]

st.dataframe(user[profile_cols].to_frame("Value"))

st.subheader("Behavior Metrics")

metric_cols = [
    "total_sessions",
    "total_watches",
    "total_watch_time",
    "avg_watch_time",
    "avg_completion",
    "recommendation_ctr",
    "recommendation_ignore_rate",
    "search_count",
    "avg_sentiment_score",
    "recommendation_trust_score",
    "content_fatigue_signal"
]

st.dataframe(user[metric_cols].to_frame("Value"))

st.subheader("Behavior Interpretation")

if user["risk_level"] == "High":
    st.warning("This user shows high engagement risk and may need re-engagement intervention.")
elif user["risk_level"] == "Medium":
    st.info("This user shows moderate risk and should be monitored.")
else:
    st.success("This user appears healthy and engaged.")
```
